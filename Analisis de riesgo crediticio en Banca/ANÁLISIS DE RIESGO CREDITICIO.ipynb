#### 1. Qué es el Riesgo de Crédito

El Riesgo de Crédito es el riesgo que asume una entidad financiera al concederle un préstamo a un cliente.

Dentro de las entidades, existen diferentes productos que tienen asociado Riesgo de Crédito, como por ejemplo:

* Créditos al consumo
* Préstamos hipotecarios
* Préstamos personales
* Tarjetas de crédito

Por otro lado, se debe de cuantificar el Riesgo de Crédito ya que el cliente puede no cumplir con su obligación de devolver el dienro en el tiempo y forma acordados. Esto puede derivar en:

* La deuda no es devuelta en su totalidad
* Se devuelve parte de la deuda
* No se devuelve la deuda en el tiempo acordado

Para reducir este riesgo, las entidades financieras suelen hacer un análisis previo, evaluando factores como el historial de pagos de la persona, sus ingresos, su nivel de endeudamiento, entro otros.

<b>¿Cómo se puede cuantificar este riesgo?</b>

Tras la crisis de 2008 los Reguladores de las diferentes geografias han generado normas a nivel nacional o comuniatrio, como en el caso de la UE, para garantizar que las entidades tengan un capital mínimo suficiente para revolver los fondos a los depositantes

Estas normas se basan en las recomendaciones de los <b>Reglamenso de Basiléa</b>. El Comité de Basilea está constituido por las autoridades de supervisión bancaria de los bancos centrales asociados. Se reúnen 4 veces al año para discutir la manera de mejorar y fortalecer el sistema bancario. Elaboran una serie de recomendaciones, que a pesar de no tener autoridad para ser cumplidas, la mayoría de los países miembros o aunque no lo sean, elaboran leyes y normas para su correcta implementación. 

En los Reglamentos de Basiela se indica que el capital mínimo suficiente debe ser calculado a través de los siguientes parámetros:

* <b>PD</b> (Probability od Default: la probabilidad de impago de un cliente
* <b>LGD</b> (Loss Given Default): mide la pérdida que se da en caso de que el cliente incurra en un impago.
* <b>EAD</b>(Exposure At Default): en caso de incumplimiento que exposición tiene el cliente en este momento

Estos parámetros pueden ser calculados por las entidades bancarias a través de modelos internos o pueden ser tomados de la regulación. Con ellos se puede cuantificar el riesgo asociado a un cliente.

En la Unión Europea para el desarrollo de modelos internos existen unas guías desarrolladas por el Banco Central Europeo donde se indican las directrices que deben cumplir los bancos al impementar modelos internos. 

Cuando un cliente solicita un producto bancario con riesgo crediticio la entidad debe calcular estos parámetros para conocer si el riesgo que va a asumir es adecuado. Estos parámetros se basan en el Scoring Crediticio que tenga el cliente.


#### 2. Modelos de Scoring Crediticio

Los modelos de Scoring Crediticio son herramientas fundamentales para la toma de decisiones en el otorgamiento de créditos por parte de las instituciones financieras.

Siempre se busca con ellos la mejor precisión en la predicción de la probabilidad de impago. Teniendo en cuenta la búsqueda del equilibrio entre minimizar el riesgo, ser competitivos y maximizar los beneficios de la entidad financiera.

Con estos modelos se busca clasificar a los clientes de forma que se pueda tomar una decisión en base al riesgo que asume la entidad al realizar el préstamo.

Los modelos deben:

* Cumplir con las leyes de la geogrofía en al que se aplqiuen en cuanto a protección de datos, inforamción crediticio y discriminación
* Se deben utilziar para su construcción datos precisos, objetivos y relevantes que midan la situación financiera y crediticia del solicitante.
* No deben discriminar por raza, género, religión o nacionalidad.
* Debe ser transparete, de modo que el solicitante pueda entender y conocer su puntuación

Con el desarrollo del modelo se pretende identificar que características de los clientes permiten predecir mejor su riesgo de crédito para poder construir una Scorecard que permita caluclar el score del solicitante dependiendo del valor de la características que se tengan en cuenta en el modelo.

#### Correlación de las variables categóricas 

Para calcular la correlación de las variables categóricas utilizamos el método de la V de Crammer que permite conocer el grado de asociación entre dos variables categóricas sin necesidad de que haya un orden o escaka.

Vamos a eliminar del dataset aquellas variables que tengan un valor en este indicador >0.4

```python
# Definimos una función para calcular la V de Crammer

def cramers_v(x,y):
    confusion_matrix = pd.crosstab(x,y)
    chi2 = stats.chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2/n
    r, k = confusion_matrix.shape
    phi2corr = max(0,phi2-((k-1)*(r-1))/(n-1))
    rcorr = r-((r-1)**2)/(n-1)
    kcorr = k-((k-1)**2)/(n-1)
    return np.sqrt(phi2corr/min((kcorr-1),(rcorr-1)))
                   
```

```python
label = preprocessing.LabelEncoder()
data_encoded = pd.DataFrame()

for i in df[categoricas+["TARGET"]].columns :
    data_encoded[i] = label.fit_transform(df[i])
rows = []

for var1 in data_encoded:
    col = []
    for var2 in data_encoded:
        cramers = cramers_v(data_encoded[var1],data_encoded[var2])
        col.append(round(cramers,2))
    rows.append(col)

cramers_results = np.array(rows)
cramer_corr = pd.DataFrame(cramers_results,columns=data_encoded.columns,index=data_encoded.columns)
cramer_corr
```

### Agrupación de las variables

Se optimiza el WOE para saber los mejroes cortes para las variables con la librería Optbinning.

Se calculan las agrupaciones para eld ataset de entrenamiento y luego se aplicarán dichas transformaciones al dataset de test.

A continuación explicamos las diferentes métricas que se van a utilziar.

* WOE (Weight Of Evidence): indica el poder predicitvo de una variable independiente en relación con la variable dependiente. Communmente se conoce como la medida de separación entre los buenos y los malos.

<b>WOE = ln(%Clientes buenos / %Clientes malos)</b>

Cuando el valor del WOE > 0 quiere decir que hay más proporción de buenos que de malos y cuando el WOE < 0 quiere decir que hay más proporcion de clientes malos que buenos.

* IV (Information Value): ayuda a clasificar las variables según su poder predictivo.

<b>IV = ∑ (%Clientes malos -%Clientes buenos) * WOE</b>

En cuanto al IV el valor indica lo siguiente: 

|IV|Prediccióm de variable|
|:----:|:----------------:|
|<0.02| No es una variable útil para la predicción|
|0,02-0.1|Poder predictivo débil|
|0.1-0.3|Poder predictivo medio|
|0.3-0.5|Fuerte poder predictivo|
|>0.5| Poder predictivo sospechoso|

```python
tar = train["TARGET"]
x = train[continuas]
woe_cont = []
iv_cont = {}
df_transform_continuas = pd.DataFrame()

for i in range(0,len(continuas)):
    woe_cont.append([])
    woe_cont[i] = OptimalBinning(name=continuas[i], dtype="numerical", solver="cp",cat_cutoff=0.1)
    woe_cont[i].fit(x.loc[:, continuas[i]], tar)
    df_transform_continuas[continuas[i]] = woe_cont[i].transform(x.loc[:, continuas[i]], metric="bins")
    woe_cont[i].binning_table.build()
    iv_cont[continuas[i]]=woe_cont[i].binning_table.iv
```

### Diseño de los modelos

Tomando de partida la misma transformación para los datos de entrada de los modelos, tomamos el dataset de entrenamiento para el ajuste del modelo que testeamos con el dataset de test-

Se van a comparar diferentes ajustes para el modelo de scoring crediticio analizadno los algoritmos de:

* Regresión Logística
* Random Forest
* Redes Neuronales
* Support Vector Machine

Cada uno de estos algoritmos cuenta con una serie de hiperparámetros que deben de ser seleccionados por el usuario antes de entrrenar el modelo. Existe una amplia combinación de métodos de búsqueda a los que indicamos varios valores para cada uno y a través de distintas técnicas encuentran la combinación óptima.

En la librería de Sickit-Leaarn las técnicas más típicas son:

* GridSearchCV: Esta técnica prubea todas las combinaciones posibles por lo que puede ser un método costoso si hay muchos hiperparámetros y valores. Por otro lado, garantiza probar todas las combinaciones y por tanto encontrará la mejor entre todas ellas.

* RandomizedSearchCV: Esta técnica selecciona combinaciones de hiperparámetros al azar, lo que reduce el tiempo de búsqueda en comparación a la técnica anterior.


En nuestro caso, vamos a utilziar la última para encontrar la combinación de los hiperparámetros que más se ajustan a nuestros datos.

De entre los parámetros que se pueden definir para esta técnica de búsqueda vamos a seleccionar los siguientes:

* estimator: El método de clasificación que se va a utilizar para ajustar el modelo.
* param_distributions: listado de los hiperparámetros con los valores para encontrar la combinación óptima, 
* n_iter: número de iteraciones
* scoring: métrica por la cual se va a seleccionar la combinación óptima. En nuestro caso utilizamos el siguiente: "balanced_accuracy", que esta indicado para datasets desbalanceados.
* n_jobs = número de jobs que se pueden ejecutar en paralelo.
* verbose= controla la cantidad de mensajes de logs. Un número más alto significa más cantidad de mensajes.
* random_state: semilla que nos permite obtener los mismos resultados si volvemos a ejecutar 
* return_train_score: indicamos que nos devuelva las métricas de entrenamiento.

#### Random Forest

Es un algoritmo de aprendizaje automático que combina múltiples árboles de decisión para realizar predicciones. Cada árbol en el bosque se entrena con una muestra aleatoria del conjunto de datos y luego se promedian las predicciones de todos los árboles para obtener una predicción final.

Es conocido por su capacidad para manejar grandes conjuntos de datos y su resistencia al sobreajuste.


En nuestro caso, estamos haciendo uso de los siguientes parámetros:

- **class_weight**: Este hiperparámetro permite asignar pesos diferentes a las clases en función de su desbalance. Puedes utilizar el valor "balanced" para que el modelo ajuste automáticamente los pesos de las clases de manera inversamente proporcional a su frecuencia en el dataset.

- **min_samples_leaf**: Este hiperparámetro establece el número mínimo de muestras requeridas para ser consideradas como una hoja. Aumentar este valor puede ayudar a evitar el sobreajuste y mejorar la generalización en un dataset desbalanceado.

- **n_estimators**: Este hiperparámetro determina el número de árboles en el bosque. Aumentar este valor puede mejorar la capacidad de generalización del modelo en un dataset desbalanceado.

- **max_features**: La cantidad de características a considerar al buscar la mejor división.

- **bootstrap**: Si se utilizan bootstrap samples al construir árboles. Si es falso, se utiliza todo el conjunto de datos para construir cada árbol.

```python
param_grid = {
    'n_estimators': [50,75,100,125,150,175,200,225,250],
    'min_samples_leaf': [int(0.025*X_train.shape[0])],
    'max_features': ['auto', 'sqrt', 'log2'],
    'bootstrap': [True, False]
}

random_search_rm = RandomizedSearchCV(
        estimator  = RandomForestClassifier(random_state=987654321, class_weight = 'balanced'),
        param_distributions = param_grid,
        n_iter     = 10, # Número máximo de combinaciones probadas
        scoring    = 'balanced_accuracy',
        n_jobs     = multiprocessing.cpu_count() - 1,
        verbose    = 0,
        random_state=987654321,
        return_train_score = True
)
random_search_rm.fit(X_train, y_train)
print(random_search_rm.best_params_)
```

### Análisis de los resultados 

Hemos seleccionado diferentes métricas para la comparación de los modelos

- **Accuracy**: indica con que frecuencia el modelo clasifica correctamente

- **Precision**: indica que porcentaje de los eventos positivos son clasificados como tal

- **Recall**: indica que porcentaje de los eventos negativos son clasificados como tal

- **AUC**: La curva AUC o ROC muestra la proporción de verdaderos positivos (el moroso se clasifica correctamente como moroso) frente a la proporción de falsos positivos (el no moroso se clasifica erróneamente como moroso).

- **Gini**: también llamado coeficiente de Gini muestra hasta qué punto el modelo tiene mejores capacidades de clasificación en comparación con el modelo aleatorio. En este contexto, el índice Gini se utiliza para evaluar la curva ROC (Receiver Operating Characteristic) del modelo, que representa la tasa de verdaderos positivos frente a la tasa de falsos positivos en función del umbral de probabilidad utilizado para hacer la clasificación. Un modelo con una curva ROC que se acerca al punto (0,1) tendrá un índice Gini más alto, lo que indica una mayor capacidad de discriminación entre las clases. Estas métrocas se relacionan de la siguiente manera:
   
                                  GINI = 2*AUC -1

- **KS**: es una métrica utilizada para evaluar la capacidad predictiva de un modelo binario. Esta métrica compara la distribución acumulada empírica de la clase positiva y la clase negativa con la distribución acumulada predicha por el modelo. Un valor alto del estadístico KS indica que el modelo es capaz de distinguir con precisión entre la clase positiva y la clase negativa, lo que significa que tiene una buena capacidad predictiva. Se calcula como la máxima diferencia entre la distribución acumulada de buenos y malos en un mismo mmento y una distribución.

                                  KS = max(abs(pa_gn - pa_bn))

              - N: número de atributos entre lod que se divide la variable
              - pa_gn: % de obserbaciones buenas acumuladas para n
              - pa_bn: % de observaciones malas acumuladas para n

### Creación de una scoredcard

Vamos a analizar que modelo nos da la representación más adecuada para la implementación de una Scorecard.

Para generar el el socore crediticio se van a tener como referencia el rango de puntos que utiliza FICO que va desde los 300 al 850.

Estos scores se dividien de la siguiente manera:

- **300-579**: son considerados "muy pobres" y representan un alto riesgo. Si la entidad decide conceder el crédito probablemnete lo haga con tasas de interés muy elevadas.
- **580-669**: represean un riesgo moderado. Es habitual que se les conceda crédito con tasas de interés elevadas.
- **670-739**: son perfiles con riesgo crediticio relativamente bajo que obtienen créditos con condiciones competitivas.
- **740-799**: representan un riesgo muy bajo, son perfiles muy buenos que consiguen las mejores tasas de interés.
- **800-850**: representan un riesgo crediticio extremadamente bajo y tienen acceso en general a las mejores condiciones para los créditos.

A continuación para poder traducir el modelo de clasificación a un score necesitamos obtener las probabilidades predichas por los modelos que calcularemos con la siguiente fórmula:
